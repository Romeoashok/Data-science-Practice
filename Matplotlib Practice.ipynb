```python
#Plot a line graph using numpy

import numpy as np

x=np.array([0,5])

y=np.array([0,50])

#plot() function is used to mark points

plt.plot(x,y)

plt.show()
```

```python
#Plot only points

import numpy as np

x=np.array([0,5])

y=np.array([0,50])

#plot() function is used to mark points

plt.plot(x,y,'o',color='red')

plt.show()
```

```python
#Plot multiple points

import numpy as np

x=np.array([0,1,2,3,4,5])

y=np.array([0,10,20,30,40,50])

#plot() function is used to mark points

plt.plot(x,y,'o',color='blue')

plt.show()
```

```python
import numpy as np

x=np.array([0,1,2,3,4,5])

y=np.array([0,10,20,30,40,50])

#plot() function is used to mark points

plt.plot(x,y,'o',marker='*',ms=30,mec='b',mfc='r')

#mec- marker edge color , mfc=marker face color

plt.show()
```

```python
days=[1,2,3,4,5,6,7]

temp1=[50,51,52,48,47,49,46]

temp2=[43,42,40,44,33,35,37]

avg=[45,48,48,46,40,42,41]

# figsize=(width,height),dpi specifies pixels per inch

plt.figure(figsize=(8,5),dpi=100)

plt.plot(days,temp1)

plt.plot(days,temp2)

plt.plot(days,avg)

plt.xlabel('mintemp')

plt.ylabel('maxtemp')

# Save figure (dpi 300 is good when saving so graph has high resolution)

plt.savefig('fig.png', dpi=300)

plt.show()
```

```python
days=[1,2,3,4,5,6,7]

temp1=[50,51,52,48,47,49,46]

temp2=[43,42,40,44,33,35,37]

avg=[45,48,48,46,40,42,41]

plt.plot(days,temp1,label="maxtemp")

plt.plot(days,temp2,label="mintemp")

plt.plot(days,avg,label="avgtemp")

plt.title("Weather")

plt.legend(loc='upper right')
```

# colormap - 'viridis'

```python
day=[1,2,3,4,5,6,7]

temp=[50,51,52,48,47,49,46]

plt.scatter(day,temp,c='red',cmap='viridis',s=30, alpha=0.5)

plt.colorbar()

plt.title('Weather')

plt.show()
```

# Bar plot

```python
day=[1,2,3,4,5,6,7]

temp=[50,51,52,48,47,49,46]

plt.bar(day,temp,color='red',width=0.5)

#Horizontal bar with height 

plt.barh(day,temp,color='red',height=0.5)

plt.title('Weather')

plt.show()
```

```python
plt.bar([1,3,5,7,9],[5,2,7,8,2], label="data1")

plt.bar([2,4,6,8,10],[8,6,2,5,6], label="data2")

plt.legend()

plt.xlabel('Bar Number')

plt.ylabel('Bar Height')

plt.title('Epic Graph')

plt.show()
```

# Pie Chart

```python
data = [10,20,30,40]

mylabels= ['color1','color2','color3','color4']

plt.pie(data,labels=mylabels,startangle=90)

plt.show()
```

```python
data = [10,20,30,40]

mylabels= ['color1','color2','color3','color4']

exp_value=[0.2, 0, 0, 0]

mycolor=['magenta', 'cyan','yellow','red']

plt.pie(data,labels=mylabels,explode=exp_value,shadow = True,colors=mycolor)

plt.legend(title = 'colors')

plt.show()
```
